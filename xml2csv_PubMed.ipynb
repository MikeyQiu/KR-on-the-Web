```python
from lxml import etree
import pandas as pd
import re
```

```python
def xml2csv(path,csv=False):
    tree = etree.parse(path)
    root = tree.getroot()
    df=pd.DataFrame()
    for child in root:
        #1 publish year
        year=child.xpath('./MedlineCitation/DateCompleted/Year/text()')[0]
        #2 title
        title=child.xpath('./MedlineCitation/Article/ArticleTitle/text()')[0]
        #3 authors
        authors='./MedlineCitation/Article/AuthorList/Author'
        names=(child.findall(authors))
        authors=[]
        for name in names:
            FN=name.findtext('ForeName')
            LN=name.findtext('LastName')
            authors.append(FN+" "+LN+",")
        author= " ".join(authors)
        #4 journal
        journal=child.xpath('./MedlineCitation/MedlineJournalInfo/MedlineTA/text()')[0]
        #print(title,authors,year,journal)
        #add new row to the dataframe
        a_row= pd.Series([title,author,year,journal])
        row_df = pd.DataFrame([a_row])
        df = pd.concat([df,row_df], ignore_index=True)
    if csv==True:
        name=" ".join(re.findall(r"(.*).xml",path))
        df.to_csv(name+'.csv')
    return df

xml2csv("test.xml",True)
```
